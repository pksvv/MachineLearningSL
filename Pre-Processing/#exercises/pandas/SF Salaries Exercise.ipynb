** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
#'chief' in 'DEPUTY chief OF DEPARTMENT'



#df['JobTitle']

def chkstring(val):
    if 'chief' in val.lower():
        return True
    else:
        return False
```

```python

sum(df['JobTitle'].apply(lambda x: chkstring(x)))
```

```python
df.loc[df['JobTitle'].str.contains('Chief')]
```
