# Merging, Joining, and Concatenation in pandas

```python
import pandas as pd
```

# Concatenation

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3']},
                  index=[0,1,2,3])
```

```python
df1.head()
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7']},
                  index=[4,5,6,7])
```

```python
df2.head()
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11']},
                  index=[8,9,10,11])
```

```python
df3.head()
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

# Merging

```python
left=pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'key':['K0','K1','K2','K3']})
```

```python
left.head()
```

```python
right = pd.DataFrame({'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3'],
                    'key':['K0','K1','K2','K3']})
```

```python
right.head()
```

```python
pd.merge(left,right,how='inner',on='key')
```

# Joining

```python
left=pd.DataFrame({'A':['A0','A1','A2'],
                    'B':['B0','B1','B2']},
                    index=['K0','K1','K2'])

right = pd.DataFrame({'C':['C0','C1','C2'],
                    'D':['D0','D1','D2']},
                    index=['K0','K2','K3'])
```

```python
right.head()
```

```python
left.join(right)
```

```python
left.join(right,how='outer')
```
