# Text Classification Assessment
This assessment is very much like the Text Classification Project we just completed, and the dataset is very similar.

The **moviereviews2.tsv** dataset contains the text of 6000 movie reviews. 3000 are positive, 3000 are negative, and the text has been preprocessed as a tab-delimited file. As before, labels are given as `pos` and `neg`.

We've included 20 reviews that contain either `NaN` data, or have strings made up of whitespace.

For more information on this dataset visit http://ai.stanford.edu/~amaas/data/sentiment/

### Task #1: Perform imports and load the dataset into a pandas DataFrame
For this exercise you can load the dataset from `'../TextFiles/moviereviews2.tsv'`.

```python
ls ..
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('TextFiles/moviereviews2.tsv',sep='\t')

df.head()
```

### Task #2: Check for missing values:

```python
# Check for NaN values:
df.isnull().sum()
```

```python
# Check for whitespace strings (it's OK if there aren't any!):

blanks = []

for i,lb,rv in df.itertuples():
    if type(rv)==str:
        if rv.isspace():
            blanks.append(i)

len(blanks)
```

### Task #3: Remove NaN values:

```python
df.dropna(inplace=True)
```

### Task #4: Take a quick look at the `label` column:

```python
df['label'].value_counts()
```

### Task #5: Split the data into train & test sets:
You may use whatever settings you like. To compare your results to the solution notebook, use `test_size=0.33, random_state=42`

```python
from sklearn.model_selection import train_test_split

x=df['review']
y=df['label']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=42)




```

### Task #6: Build a pipeline to vectorize the date, then train and fit a model
You may use whatever model you like. To compare your results to the solution notebook, use `LinearSVC`.

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC



text_clf_svc = Pipeline([('tfidf',TfidfVectorizer()),('clf',LinearSVC())])



text_clf_svc.fit(x_train,y_train)
```

### Task #7: Run predictions and analyze the results

```python
# Form a prediction set
predictions = text_clf_svc.predict(x_test)
```

```python
# Report the confusion matrix
from sklearn import metrics
print('\n Confusion Metrics \n',metrics.confusion_matrix(y_test,predictions))
print('\n Classification Report \n',metrics.classification_report(y_test,predictions))
print('\n Accuracy Metrics \n',metrics.accuracy_score(y_test,predictions))
```
